```python
# Importacion de librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.metrics import silhouette_score, calinski_harabasz_score
```

```python
# Cargar el archivo CSV desde la carpeta proporcionada
file_path = './DataSet/Mall_Customers.csv'
data = pd.read_csv(file_path)

# Exploraci√≥n de datos
print(data.info())
print(data.describe())
```

```python
# Gr√°fica de distribuci√≥n de ingresos anuales y puntuaci√≥n de gasto
sns.scatterplot(x=data['Annual Income (k$)'], y=data['Spending Score (1-100)'])
plt.title('Distribuci√≥n de Ingresos vs Puntuaci√≥n de Gasto')
plt.xlabel('Ingreso Anual (k$)')
plt.ylabel('Puntuaci√≥n de Gasto')
plt.show()
```

```python
# Identificaci√≥n y tratamiento de valores at√≠picos

# Funci√≥n para identificar valores at√≠picos usando el rango intercuartil
def detectar_outliers(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[columna] < lower_bound) | (df[columna] > upper_bound)]

# Detectar valores at√≠picos en las variables relevantes
outliers_annual_income = detectar_outliers(data, 'Annual Income (k$)')
outliers_spending_score = detectar_outliers(data, 'Spending Score (1-100)')

# Mostrar los valores at√≠picos detectados
print("Valores at√≠picos en Ingreso Anual:")
print(outliers_annual_income)
print("\nValores at√≠picos en Puntuaci√≥n de Gasto:")
print(outliers_spending_score)

# Decisi√≥n: eliminar los valores at√≠picos detectados
data = data[~data.index.isin(outliers_annual_income.index)]
data = data[~data.index.isin(outliers_spending_score.index)]
```

```python
# Selecci√≥n de caracter√≠sticas y escalado
features = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Dendrograma
linked = linkage(scaled_features, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(linked, truncate_mode='lastp', p=12, leaf_rotation=45., leaf_font_size=10., show_contracted=True)
plt.title('Dendrograma para determinar clusters √≥ptimos')
plt.xlabel('Muestras (Clientes)')
plt.ylabel('Distancia Euclidiana')
plt.show()
```

```python
# N√∫mero √≥ptimo de clusters seg√∫n el dendrograma
n_clusters = 5

# Aplicar Hierarchical Clustering
clusters = fcluster(linked, n_clusters, criterion='maxclust')

# Agregar los clusters al DataFrame original
data['Cluster'] = clusters

# Visualizaci√≥n de resultados
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x='Annual Income (k$)', y='Spending Score (1-100)',
    hue='Cluster', data=data, palette='Set1'
)
plt.title('Clusters obtenidos con Hierarchical Clustering')
plt.xlabel('Ingreso Anual (k$)')
plt.ylabel('Puntuaci√≥n de Gasto')
plt.legend(title='Cluster')
plt.show()

# Resumen de los clusters
cluster_summary = data.groupby('Cluster').agg({
    'Age': ['mean', 'std', 'min', 'max'],
    'Annual Income (k$)': ['mean', 'std', 'min', 'max'],
    'Spending Score (1-100)': ['mean', 'std', 'min', 'max']
}).reset_index()

# Mostrar resumen
print(cluster_summary)
```

```python
# Evaluaci√≥n del modelo
scaled_data = scaler.transform(data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
final_silhouette = silhouette_score(scaled_data, data['Cluster'])
final_calinski = calinski_harabasz_score(scaled_data, data['Cluster'])

print("\nEvaluaci√≥n del modelo Hierarchical Clustering:")
print(f"Silhouette Score: {final_silhouette}")
print(f"Calinski-Harabasz Score: {final_calinski}")
```
